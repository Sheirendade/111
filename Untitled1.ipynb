```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_model import ARMA

np.random.seed(0)
```

# Construct the data

```python
df=pd.read_csv('./sp500_index.csv')
```

```python
daily_return=[]

for i in range(1,df.shape[0]):
  a=df.iloc[i]['S&P500']
  b=df.iloc[i-1]['S&P500']
  c=(a-b)/b
  daily_return.append(c)
```

```python
Daily_return=pd.DataFrame(data=daily_return, columns=['value'], dtype=float)
Daily_return.value=Daily_return.value*100
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 5))

plt.plot(Daily_return,label='initial data')
plt.xlim([0, 200])

ax.set_xlabel('Date')
ax.set_ylabel('SP500')
ax.grid(True)
ax.legend(loc='upper right');

```

```python
y=Daily_return.value
adf_result = adfuller(y)
print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' % adf_result[1])
```

The p-value is less than 0.05. Thus, we can reject the null hypothesis (H0). The time series does not have a unit root and is stationary. 

# ACF and PACF Plots

```python
# plot acf and pacf
plot_pacf(y)
plot_acf(y)
```

# Choose the best model

```python
# lists of candidate orders
MA_orders = list(range(0, 5))
AR_orders = list(range(0, 5))

# a new dataframe to store the results (AIC)
results = pd.DataFrame(columns = ['AR order', 'MA order', 'AIC'])
```

```python
# fit ARMA model
for AR in AR_orders:
    for MA in MA_orders:
        if AR + MA != 0:
            try:
                result = ARMA(y, order = (AR, MA)).fit()
                results = results.append({'AR order': AR, 'MA order': MA, 'AIC': result.aic}, ignore_index = True)
            except: # ValueError: The computed initial MA coefficients are not invertible
                print('AR: %i, MA: %d failed' % (AR, MA))
```

```python
# sort models by AIC
results = results.sort_values(by = ['AIC'], ascending = True)
# display the results
results




```

According to AIC, ARMA(4,4) fits the SP-500 time series best. However, we observe that the AIC of the ARMA(2,2) process is pretty close to that of ARMA(4,4). Based on the parsimony principle, we choose ARMA(2,2) process.

```python
#@title
# estimate the parameters
arma_22 = ARMA(y, order = (2, 2)).fit()

# print the summary
arma_22.summary()




```

```python
# in sample
time_series = pd.DataFrame(columns = ['initial_data', 'prediction'])
time_series['initial_data'] = y
time_series['prediction'] = arma_22.fittedvalues
time_series['residual'] = time_series['initial_data'] - time_series['prediction']
```

```python
# plot the first 200 data points
time_series.loc[:200, ['initial_data', 'prediction']].plot(figsize=(10,5))
```

```python
#@title
# plot the first 200 residuals
time_series.loc[:200, ['residual']].plot(figsize=(10,5))
```
